```python
# Load weather daily data
weather_data = pd.read_csv('weather_daily_darksky.csv')

# Inspect the first few rows and data info
print(weather_data.head())
print(weather_data.info())

```

```python
# Convert the 'time' column to datetime
weather_data['time'] = pd.to_datetime(weather_data['time'])

# Check and fill missing values
weather_data.fillna(method='ffill', inplace=True)  # Forward fill for missing values

```

```python
# Merge weather data with daily energy data
combined_data = pd.merge(daily_data, weather_data, left_on='day', right_on='time', how='left')
```

```python
# Correlation analysis
correlation = combined_data[['energy_sum', 'temperatureMax', 'humidity', 'windSpeed', 'cloudCover']].corr()
print(correlation)
```

```python
combined_data.plot(x='temperatureMax', y='energy_sum', kind='scatter', alpha=0.5)
plt.title("Energy Consumption vs Maximum Temperature")
plt.xlabel("Temperature Max (°C)")
plt.ylabel("Energy Consumption (kWh)")
plt.show()
```

```python
combined_data.set_index('day')[['energy_sum', 'humidity']].plot(figsize=(12, 6))
plt.title("Energy Consumption and Humidity Over Time")
plt.show()
```

```python
# Add seasons
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

combined_data['season'] = combined_data['day'].apply(get_season)

# Group by season and calculate mean energy consumption
seasonal_energy = combined_data.groupby('season')['energy_sum'].mean()
seasonal_energy.plot(kind='bar', color='skyblue')
plt.title("Average Energy Consumption by Season")
plt.xlabel("Season")
plt.ylabel("Energy Consumption (kWh)")
plt.show()
```

```python
# Create temperature bins
combined_data['temp_bins'] = pd.cut(combined_data['temperatureMax'], bins=[0, 5, 10, 15, 20], labels=['0-5°C', '5-10°C', '10-15°C', '15-20°C'])

# Group by temperature bins and calculate average energy consumption
temp_avg_energy = combined_data.groupby('temp_bins')['energy_sum'].mean()

# Plot the aggregated data
temp_avg_energy.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title("Average Energy Consumption by Temperature Range")
plt.xlabel("Temperature Range")
plt.ylabel("Average Energy Consumption (kWh)")
plt.show()

```

```python
import numpy as np
from scipy.stats import linregress

# Fit a linear trend line
x = combined_data['temperatureMax']
y = combined_data['energy_sum']
mask = ~np.isnan(x) & ~np.isnan(y)  # Remove NaNs
slope, intercept, r_value, p_value, std_err = linregress(x[mask], y[mask])

# Plot the data and trend line
plt.scatter(x, y, alpha=0.3, label="Data")
plt.plot(x, slope * x + intercept, color='red', label=f"Trend Line (R²={r_value**2:.2f})")
plt.title("Energy Consumption vs Temperature (Simplified)")
plt.xlabel("Temperature Max (°C)")
plt.ylabel("Energy Consumption (kWh)")
plt.legend()
plt.show()
```

```python
# Sort data by temperature and calculate a rolling mean
combined_data = combined_data.sort_values('temperatureMax')
combined_data['rolling_mean'] = combined_data['energy_sum'].rolling(window=50).mean()

# Plot smoothed data
plt.plot(combined_data['temperatureMax'], combined_data['rolling_mean'], color='blue')
plt.title("Smoothed Energy Consumption vs Temperature")
plt.xlabel("Temperature Max (°C)")
plt.ylabel("Energy Consumption (kWh)")
plt.show()
```

```python
# Load the datasets
households_info = pd.read_csv('informations_households.csv')
energy_data = pd.read_csv('daily_dataset.csv')

# Merge the datasets on the 'LCLid' column
merged_data = pd.merge(energy_data, households_info, on='LCLid')

# Filter for relevant ACORN groups
filtered_data = merged_data[merged_data['Acorn_grouped'].isin(['Affluent', 'Adversity', 'Comfortable'])]

# Group by 'Acorn_grouped' and calculate the total energy consumption
energy_by_group = filtered_data.groupby('Acorn_grouped')['energy_sum'].sum()

# Plot the energy consumption by ACORN group
energy_by_group.plot(kind='bar', figsize=(8, 5), color=['blue', 'orange', 'green'])
plt.title("Total Energy Consumption by ACORN Group")
plt.xlabel("ACORN Group")
plt.ylabel("Total Energy Consumption (kWh)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
